```python
# importing libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
# reading data
X_full = pd.read_csv('Data/train.csv', index_col='Id')
X_test_full = pd.read_csv('Data/test.csv', index_col='Id')
```

```python
# examining training data set
X_full.head()
```

```python
# examining test data set
X_test_full.head()
```

```python
# setting 'SalePrice' column as target for predctions
y = X_full.SalePrice

# setting feauters for predictions
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']

# making deep copy of DataFrame features object (training and test)
X = X_full[features].copy()
X_test = X_test_full[features].copy()
```

```python
# splitting X, y datasets for training and validating subsets
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
# examining X_train dataset
X_train.head()
```

```python
# defining five models with five different parameter values  for evaluating the best available model
model_1 = RandomForestRegressor(n_estimators=50, random_state=0)
model_2 = RandomForestRegressor(n_estimators=100, random_state=0)
model_3 = RandomForestRegressor(n_estimators=100, criterion='mae', random_state=0)
model_4 = RandomForestRegressor(n_estimators=200, min_samples_split=20, random_state=0)
model_5 = RandomForestRegressor(n_estimators=100, max_depth=7, random_state=0)

# creating models list
models = [model_1, model_2, model_3, model_4, model_5]
```

```python
# evaluating the best prediction model
def score_model(model, X_t=X_train, X_v=X_valid, y_t=y_train, y_v=y_valid):
    #fit data to model
    model.fit(X_t, y_t)
    
    #predictions
    preds = model.predict(X_v)
    
    return mean_absolute_error(y_v, preds)

# comparing models' results in loop
mae_list = [0,0,0,0,0]
for i in range(0, len(models)):
    mae_list[i] = score_model(models[i])
    
```

```python
# selecting smallest mae value
min(mae_list)
```
